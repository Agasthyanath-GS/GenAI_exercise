```python
from openai import OpenAI

openai_api_key = os.getenv("OPENAI_API_KEY")
openai_api_key = os.getenv("OPENAI_API_KEY")
openai_client = OpenAI(api_key=openai_api_key)
```

```python
messages = [{"role":"user","content":"Hello, how are you?"}]
```

```python
response = openai_client.chat.completions.create(model="gpt-5-nano", messages=messages)
```

```python
print(response.choices[0].message.content)
```

```python

question = "create a hard challenging question to access someone's IQ, respond only with the question, strictly do not add any other sentence or words"
message = [{"role": "user","content": question}]
```
