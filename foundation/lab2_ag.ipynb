### Import required Packages

```python
# import the required packages

import os
import json
from openai import OpenAI
from groq import Groq
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

### 1. setting up the env and api clients

```python
# load the api keys
load_dotenv(override=True)
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
GROQ_API_KEY   = os.getenv("GROQ_API_KEY")
```

```python
# message for llm's
question  ="Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence."
question += "only provide the question,strictly do not add any other explanations"
message   = [{"role":"user", "content":f"{question}"}]
message
```

```python
# setting the api clients 
openai_client = OpenAI(api_key=OPENAI_API_KEY)
gemini_client = OpenAI(api_key=GOOGLE_API_KEY, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
groq_client   = OpenAI(api_key=GROQ_API_KEY, base_url="https://api.groq.com/openai/v1")
```

```python
# calling open ai gpt 4o mini for generating the question.
model ="gpt-4o-mini"
response = openai_client.chat.completions.create(model=model,messages=message)
question = response.choices[0].message.content
display(Markdown(question))
```

### 2. Compairing response from multiple models

```python
# save the responses into a list for further checkings.
responses  = []
model_name = []
message = [{"role":"user","content":f"{question}"}]
print(message)
```

```python
model ="gpt-4o-mini"
response = openai_client.chat.completions.create(model=model,messages=message)
answer = response.choices[0].message.content
display(Markdown(answer))
responses.append(answer)
model_name.append(model)
```

```python
# compairing multiple models
model ="qwen/qwen3-32b"
response = groq_client.chat.completions.create(model=model,messages=message)
answer = response.choices[0].message.content
display(Markdown(answer))
responses.append(answer)
model_name.append(model)
```

```python
model ="gpt-5-nano"
response = openai_client.chat.completions.create(model=model,messages=message,)
answer = response.choices[0].message.content
display(Markdown(answer))
responses.append(answer)
model_name.append(model)
```

```python
model ="llama-3.3-70b-versatile"
response = groq_client.chat.completions.create(model=model,messages=message)
answer = response.choices[0].message.content
display(Markdown(answer))
responses.append(answer)
model_name.append(model)
```

### 3. Finding the Best Response from all.

```python
command = f""" consider you as a judge and now you are juding the response from {len(responses)} competitiors.
the response recieved from the competitiors for the question : {question}, 
Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{responses}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks.
"""
```

```python
judge_command = [{"role":"user", "content":f"{command}"}]
```

```python
model ="gpt-4o-mini"
response = openai_client.chat.completions.create(model=model,messages=judge_command)
judgement = response.choices[0].message.content
display(Markdown(judgement))
```

```python
results = json.loads(judgement)
result_list = list(results.values())[0]
for i,model in enumerate(model_name):
    print(f"-> rank {i+1} : {model_name[int(result_list[i])-1]}")
```
