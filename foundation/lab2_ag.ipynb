```python
# import the required packages

import os
import json
from openai import OpenAI
from groq import Groq
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

```python
# load the api keys
load_dotenv(override=True)
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
GROQ_API_KEY   = os.getenv("GROQ_API_KEY")
```

```python
# message for llm's
question  ="Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence."
question += "only provide the question,strictly do not add any other explanations"
message   = [{"role":"user", "content":f"{question}"}]
message
```

```python
# setting the api clients 
openai_client = OpenAI(api_key=OPENAI_API_KEY)
gemini_client = OpenAI(api_key=GOOGLE_API_KEY, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
groq_client   = OpenAI(api_key=GROQ_API_KEY, base_url="https://api.groq.com/openai/v1")
```

```python
# calling open ai gpt 4o mini for generating the question.
model ="gpt-4o-mini"
response = openai_client.chat.completions.create(model=model,messages=message)
question = response.choices[0].message.content
display(Markdown(question))
```

### Compairing response from multiple models

```python
# save the responses into a list for further checkings.
responses  = []
model_name = []
message = [{"role":"user","content":f"{question}"}]
print(message)
```
