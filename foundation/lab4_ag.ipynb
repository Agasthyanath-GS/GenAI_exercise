### Importing Required Library and Packages

```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
```

### Checking and setting up PushOver

```python
load_dotenv(override=True)
PUSHOVER_KEY = os.getenv(key="PUSHOVER_KEY")
PUSHOVER_TOKEN = os.getenv(key="PUSHOVER_TOKEN")
PUSHOVER_URL = "https://api.pushover.net/1/messages.json"
```

```python
def push(message):
    print(f"Push: {message}")
    payload = {"user": PUSHOVER_KEY, "token": PUSHOVER_TOKEN, "message": message}
    requests.post(PUSHOVER_URL, data=payload)
```

```python
push("running")
```
